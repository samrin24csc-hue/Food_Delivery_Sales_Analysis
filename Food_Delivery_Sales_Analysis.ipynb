```python
import pandas as pd
import numpy as np

np.random.seed(42)

food_items = ['Burger', 'Pizza', 'Biryani', 'Pasta', 'Sandwich']
locations = ['Chennai', 'Madurai', 'Trichy', 'Coimbatore', 'Salem']
times = ['10 AM', '12 PM', '2 PM', '4 PM', '6 PM', '8 PM']

data = {
    'Order_ID': range(1, 201),
    'Food_Item': np.random.choice(food_items, 200),
    'Location': np.random.choice(locations, 200),
    'Order_Time': np.random.choice(times, 200),
    'Quantity': np.random.randint(1, 5, 200),
    'Price': np.random.randint(100, 500, 200)
}

df = pd.DataFrame(data)

df['Revenue'] = df['Quantity'] * df['Price']

df.to_csv('food_delivery_data.csv', index=False)

df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nColumn Information:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

food_count = df['Food_Item'].value_counts()

food_count.plot(kind='bar')
plt.title("Most Popular Food Items")
plt.xlabel("Food Item")
plt.ylabel("Orders")
plt.show()
```

```python
revenue_food = df.groupby('Food_Item')['Revenue'].sum()

revenue_food.plot(kind='bar')
plt.title("Revenue by Food Item")
plt.ylabel("Revenue")
plt.show()
```

```python
location_revenue = df.groupby('Location')['Revenue'].sum()

location_revenue.plot(kind='bar')
plt.title("Revenue by Location")
plt.ylabel("Revenue")
plt.show()
```

```python
time_orders = df['Order_Time'].value_counts()

time_orders.plot(kind='line', marker='o')
plt.title("Orders by Time")
plt.xlabel("Order Time")
plt.ylabel("Number of Orders")
plt.show()
```

```python
revenue_food.plot(kind='pie', autopct='%1.1f%%')
plt.title("Revenue Distribution")
plt.ylabel('')
plt.show()
```

```python
print("BUSINESS INSIGHTS")

print("\n1. Most Popular Food Item:")
print(df['Food_Item'].value_counts().idxmax())

print("\n2. Highest Revenue Food Item:")
print(df.groupby('Food_Item')['Revenue'].sum().idxmax())

print("\n3. Highest Revenue Location:")
print(df.groupby('Location')['Revenue'].sum().idxmax())

print("\n4. Peak Order Time:")
print(df['Order_Time'].value_counts().idxmax())
```
